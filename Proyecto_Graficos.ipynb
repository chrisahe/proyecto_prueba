## Importamos Librerias

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Leemos la base de datos .csv desde drive


```python
path= '/content/drive/MyDrive/Análisis de Datos/Bases datos Proyecto/proyecto_clean.csv'
df = pd.read_csv(path)
```

##  Verificamos data frame

```python
df
```

## Pie Chart o gráfico de torta (%)

```python
df_elect2 = df[df['COMBUSTIBLE'] == 'ELECTRICO']

# Agrupamos y calculamos porcentajes
resultado2 = df_elect2.groupby('DEPARTAMENTO')['COMBUSTIBLE'].count()
porcentajes = resultado2 / resultado2.sum() * 100  # Calculamos porcentajes

# Creamos el gráfico de torta sin etiquetas en las fracciones
resultado2.plot(kind='pie', labels=None,  # Sin etiquetas en las fracciones
               figsize=(10, 10),
               wedgeprops={'width': 0.3},
               ylabel='', fontsize=8)

# Creamos la leyenda con porcentajes
leyenda_etiquetas = [f'{dep} ({porcentaje:.2f}%)' for dep, porcentaje in zip(resultado2.index, porcentajes)]
plt.legend(fontsize=10, labels=leyenda_etiquetas, loc='upper right')

plt.tight_layout()
plt.title('Porcentaje de vehiculos ').set_fontsize(20)
plt.show()
```

```python
df_elect2 = df[df['COMBUSTIBLE'] == 'ELECTRICO']

# Agrupamos y calculamos porcentajes
resultado2 = df_elect2.groupby('DEPARTAMENTO')['COMBUSTIBLE'].count()
porcentajes = resultado2 / resultado2.sum() * 100  # Calculamos porcentajes

# Creamos el gráfico de torta sin etiquetas en las fracciones
resultado2.plot(kind='pie', labels=None,  # Sin etiquetas en las fracciones
               figsize=(10, 10),
               wedgeprops={'width': 0.3},
               ylabel='', fontsize=8)

# Creamos la leyenda con porcentajes y la movemos fuera del gráfico
leyenda_etiquetas = [f'{dep} ({porcentaje:.2f}%)' for dep, porcentaje in zip(resultado2.index, porcentajes)]
plt.legend(fontsize=10, labels=leyenda_etiquetas, loc='upper left', bbox_to_anchor=(1.05, 1))

plt.tight_layout()
plt.title('Porcentaje de vehiculos ').set_fontsize(20)
plt.show()
```
