```python
tendencia_combustible = proyecto.groupby('AÑO_REGISTRO')['COMBUSTIBLE'].count().reset_index()

plt.figure(figsize=(5, 3))
sns.lineplot(x='AÑO_REGISTRO', y='COMBUSTIBLE', data = tendencia_combustible, marker='o')
plt.title('Tendencia Tipo de Combustible')
plt.xlabel('AÑO_REGISTRO')
plt.ylabel('COMBUSTIBLE')
plt.grid(True)
plt.show()
```

```python
tendencia_combustible2 = proyecto.groupby(['AÑO_REGISTRO', 'COMBUSTIBLE'])['CANTIDAD'].sum().unstack()

tendencia_combustible2.plot(kind='bar', stacked=True, figsize=(12, 8))
plt.title('Tendencia Combustible Anual')
plt.xlabel('AÑO_REGISTRO')
plt.ylabel('CANTIDAD_VEHICULOS')
plt.legend(title='Tipo de Combustible')
plt.grid(True)
plt.show()
```
