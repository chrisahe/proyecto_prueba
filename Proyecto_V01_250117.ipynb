###Eliminar variables que no se requieren en el dataframe (null)

```python
vh_electricos_hibridos.drop("CAPACIDAD_PASAJEROS",axis=1, inplace=True)
vh_electricos_hibridos.drop("CAPACIDAD_CARGA",axis=1, inplace=True)
vh_electricos_hibridos
```

###Gráficos

```python
from matplotlib import pyplot as plt
import seaborn as sns
figsize = (12, 1.2 * len(vh_electricos_hibridos['COMBUSTIBLE'].unique()))
plt.figure(figsize=figsize)
sns.violinplot(vh_electricos_hibridos, x='AÑO_REGISTRO', y='COMBUSTIBLE', inner='box', palette='Dark2')
sns.despine(top=True, right=True, bottom=True, left=True)
```

```python
tendencia_combustible = vh_electricos_hibridos.groupby(['AÑO_REGISTRO', 'COMBUSTIBLE'])['CANTIDAD'].sum().unstack()

tendencia_combustible.plot(kind='bar', stacked=True, figsize=(5, 5))
plt.title('TENDENCIA DE REGISTRO X TIPO DE COMBUSTIBLE')
plt.xlabel('AÑO_REGISTRO')
plt.ylabel('CANTIDAD_VEHICULOS')
plt.legend(title='TIPO_COMBUSTIBLE')
plt.grid(True)
plt.show()
```

```python
vh_electricos_hibridos_2 = vh_electricos_hibridos[vh_electricos_hibridos['COMBUSTIBLE'] == 'ELECTRICO']

# Agrupamos y calculamos porcentajes
resultado2 = vh_electricos_hibridos_2.groupby('DEPARTAMENTO')['COMBUSTIBLE'].count()
porcentajes = resultado2 / resultado2.sum() * 100  # Calculamos porcentajes

# Creamos el gráfico de torta sin etiquetas en las fracciones
resultado2.plot(kind='pie', labels=None,  # Sin etiquetas en las fracciones
               figsize=(10, 10),
               wedgeprops={'width': 0.3},
               ylabel='', fontsize=8)

# Creamos la leyenda con porcentajes y la movemos fuera del gráfico
leyenda_etiquetas = [f'{dep} ({porcentaje:.2f}%)' for dep, porcentaje in zip(resultado2.index, porcentajes)]
plt.legend(fontsize=10, labels=leyenda_etiquetas, loc='upper left', bbox_to_anchor=(1.05, 1))

plt.tight_layout()
plt.title('Porcentaje de vehiculos ').set_fontsize(20)
plt.show()
```
